```python
import pandas as pd
df=pd.read_csv('./vg-stats/vgsales.csv')

df
```

# Which company is the most common video game publisher?

```python
most_game_publisher= df['Publisher'].mode()
most_game_publisher[0]
```

# What’s the most common platform?

```python
most_common_platform=df['Platform'].mode()
most_common_platform[0]
```

# What about the most common genre?

```python
most_common_genre=df['Genre'].mode()
most_common_genre[0]
```

# What are the top 20 highest grossing games?


```python
top_20=df[["Name","Global_Sales"]].sort_values('Global_Sales',ascending=False)
top_20.head(20)
```

# For North American video game sales, what’s the median?


```python
NA_median=df['NA_Sales'].median()
NA_median
bigger=df[df["NA_Sales"]>NA_median][['NA_Sales','Name']].sort_values('NA_Sales').head(5)
bigger
less=df[df["NA_Sales"]<NA_median][['NA_Sales','Name']].sort_values('NA_Sales').tail(5)
less
compine=[bigger,less]
sorounding=pd.concat(compine)
sorounding
```

# For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
top_20=df.sort_values('Global_Sales',ascending=False)
top_20.head(20)
```

# The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?


```python
wii_avg = df[df['Platform']=='Wii'].Global_Sales.mean()
others_avg = df[df['Platform']!='Wii'].Global_Sales.mean()
others_avg , wii_avg
```

# what is the top 5 most seling game in europ 

```python
most_seling=df[['Name','EU_Sales']].sort_values('EU_Sales', ascending=False).set_index('Name').head()
most_seling
```
