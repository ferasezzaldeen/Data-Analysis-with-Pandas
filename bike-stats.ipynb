```python
import pandas as pd
df_station= pd.read_csv('./bike-state/station.csv')
df_trip= pd.read_csv('./bike-state/trip.csv', error_bad_lines=False)
df_weather= pd.read_csv('./bike-state/weather.csv')
df_station
# df_trip
df_weather
```

# What is the average trip duration for a borrowed bicycle?

```python
avg_duration=df_trip["tripduration"].mean()
avg_duration
```

# What’s the most common age of a bicycle-sharer?

```python
avg_bd=df_trip["birthyear"].mode()
common_age=2021-avg_bd
common_age
```

# Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
df_weather["month"]=pd.DatetimeIndex(df_weather["Date"]).month
df_weather["year"]=pd.DatetimeIndex(df_weather["Date"]).year
avg_preciption=df_weather[['Precipitation_In',"month",'year']].groupby(['year','month']).mean()
median_preciption=df_weather[['Precipitation_In',"month",'year']].groupby(['year','month']).median()
avg_preciption, median_preciption
```
